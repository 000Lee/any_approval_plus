```python
# ===========================================
# 1. ë°ì´í„° ë¡œë“œ (CSV + DB)
# ===========================================
import pymysql
import pandas as pd
import json

# CSV ë¡œë“œ
in_path = r"C:\Users\LEEJUHWAN\Desktop\ì• ë‹ˆíŒŒì´ë¸Œ\ë°±ì—…\in_employee.csv"
out_path = r"C:\Users\LEEJUHWAN\Desktop\ì• ë‹ˆíŒŒì´ë¸Œ\ë°±ì—…\out_employee.csv"

df_in = pd.read_csv(in_path, encoding='utf-8-sig')
df_out = pd.read_csv(out_path, encoding='utf-8-sig')

# ë”•ì…”ë„ˆë¦¬ ìƒì„±
in_dict = {}
for _, row in df_in.iterrows():
    name = str(row['ì‚¬ì›ëª…']).strip()
    in_dict[name] = {
        'emailId': str(row['ID']).strip() if pd.notna(row['ID']) else '',
        'deptName': str(row['ë¶€ì„œ']).strip() if pd.notna(row['ë¶€ì„œ']) else '',
        'positionName': str(row['ì§ìœ„']).strip() if pd.notna(row['ì§ìœ„']) else '',
        'deptCode': str(row['ë¶€ì„œì½”ë“œ']).strip() if pd.notna(row['ë¶€ì„œì½”ë“œ']) else ''
    }

out_dict = {}
for _, row in df_out.iterrows():
    name = str(row['ì‚¬ì›ëª…']).strip()
    out_dict[name] = {
        'emailId': str(row['ID']).strip() if pd.notna(row['ID']) else '',
        'deptName': str(row['ë¶€ì„œ']).strip() if pd.notna(row['ë¶€ì„œ']) else '',
        'positionName': str(row['ì§ìœ„']).strip() if pd.notna(row['ì§ìœ„']) else '',
        'deptCode': str(row['ë¶€ì„œì½”ë“œ']).strip() if pd.notna(row['ë¶€ì„œì½”ë“œ']) else ''
    }

# DB ì¡°íšŒ (referrers ì»¬ëŸ¼ ì¶”ê°€)
conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    password='1234',
    database='any_approval',
    charset='utf8mb4'
)

query = """SELECT id, drafter_name, drafter_email, drafter_position, 
           drafter_dept, drafter_dept_code, activities, referrers FROM documents"""
df_db = pd.read_sql(query, conn)
conn.close()

print(f"ì¬ì§ì CSV: {len(in_dict)}ëª…")
print(f"í‡´ì‚¬ì CSV: {len(out_dict)}ëª…")
print(f"DB documents: {len(df_db)}ê±´")
```

```python
# ===========================================
# 2. ê³µí†µ í•¨ìˆ˜ ì •ì˜
# ===========================================
def get_expected_info(name):
    """ì´ë¦„ìœ¼ë¡œ ì˜ˆìƒ ì •ë³´ ë°˜í™˜"""
    name = str(name).strip() if name else ''
    if name in in_dict:
        return in_dict[name], 'in'
    if name in out_dict:
        return out_dict[name], 'out'
    return {'emailId': '', 'deptName': '', 'positionName': '', 'deptCode': ''}, 'unknown'

def safe_str(value):
    """nan ì²˜ë¦¬ëœ ë¬¸ìì—´ ë°˜í™˜"""
    if pd.isna(value) or str(value).strip() == 'nan':
        return ''
    return str(value).strip()
```

```python
# ===========================================
# 3. Drafter ê²€ì¦ + Unknown ì§‘ê³„
# ===========================================
drafter_errors = []
drafter_ok = 0
unknown_drafters = []  # Unknown ì¼€ì´ìŠ¤ ì§‘ê³„

for _, row in df_db.iterrows():
    name = row['drafter_name']
    expected, source = get_expected_info(name)
    
    # Unknown ì¼€ì´ìŠ¤ ì§‘ê³„
    if source == 'unknown':
        unknown_drafters.append({'id': row['id'], 'name': name})
    
    # ì‹¤ì œ DB ê°’
    actual_email = safe_str(row['drafter_email'])
    actual_dept = safe_str(row['drafter_dept'])
    actual_position = safe_str(row['drafter_position'])
    actual_dept_code = safe_str(row['drafter_dept_code'])
    
    # ì˜ˆìƒ ê°’
    exp_email = safe_str(expected['emailId'])
    exp_dept = safe_str(expected['deptName'])
    exp_position = safe_str(expected['positionName'])
    exp_dept_code = safe_str(expected['deptCode'])
    
    # ë¹„êµ
    errors = []
    if actual_email != exp_email:
        errors.append(f"email: DB '{actual_email}' vs CSV '{exp_email}'")
    if actual_dept != exp_dept:
        errors.append(f"dept: DB '{actual_dept}' vs CSV '{exp_dept}'")
    if actual_position != exp_position:
        errors.append(f"position: DB '{actual_position}' vs CSV '{exp_position}'")
    if actual_dept_code != exp_dept_code:
        errors.append(f"dept_code: DB '{actual_dept_code}' vs CSV '{exp_dept_code}'")
    
    if errors:
        drafter_errors.append({
            'id': row['id'],
            'name': name,
            'source': source,
            'errors': errors
        })
    else:
        drafter_ok += 1

print("=" * 60)
print("Drafter ê²€ì¦ ê²°ê³¼")
print("=" * 60)
print(f"âœ… ì •ìƒ: {drafter_ok}ê±´")
print(f"âŒ ì˜¤ë¥˜: {len(drafter_errors)}ê±´")

# Unknown ì¼€ì´ìŠ¤ ì¶œë ¥
unique_unknown = set([u['name'] for u in unknown_drafters])
print(f"\nâš ï¸  CSVì— ì—†ëŠ” ê¸°ì•ˆì (unknown): {len(unique_unknown)}ëª…")
if unique_unknown:
    print(f"   ì´ë¦„ ëª©ë¡: {unique_unknown}")

if drafter_errors:
    print("\n[ì˜¤ë¥˜ ìƒ˜í”Œ (ìµœëŒ€ 10ê°œ)]")
    for err in drafter_errors[:10]:
        print(f"  ID {err['id']}: {err['name']} ({err['source']})")
        for e in err['errors']:
            print(f"    - {e}")
```

```python
# ===========================================
# 4. Activities ê²€ì¦ + Unknown ì§‘ê³„
# ===========================================
activities_errors = []
activities_ok = 0
activities_total = 0
unknown_activities = []  # Unknown ì¼€ì´ìŠ¤ ì§‘ê³„

for _, row in df_db.iterrows():
    if not row['activities'] or pd.isna(row['activities']):
        continue
    
    try:
        activities = json.loads(row['activities'])
    except:
        continue
    
    for act in activities:
        activities_total += 1
        name = str(act.get('name', '')).strip()
        expected, source = get_expected_info(name)
        
        # Unknown ì¼€ì´ìŠ¤ ì§‘ê³„
        if source == 'unknown':
            unknown_activities.append({'doc_id': row['id'], 'name': name})
        
        # ì‹¤ì œ ê°’
        actual_email = safe_str(act.get('emailId', ''))
        actual_dept = safe_str(act.get('deptName', ''))
        actual_position = safe_str(act.get('positionName', ''))
        actual_dept_code = safe_str(act.get('deptCode', ''))
        
        # ì˜ˆìƒ ê°’
        exp_email = safe_str(expected['emailId'])
        exp_dept = safe_str(expected['deptName'])
        exp_position = safe_str(expected['positionName'])
        exp_dept_code = safe_str(expected['deptCode'])
        
        # ë¹„êµ
        errors = []
        if actual_email != exp_email:
            errors.append(f"emailId: DB '{actual_email}' vs CSV '{exp_email}'")
        if actual_dept != exp_dept:
            errors.append(f"deptName: DB '{actual_dept}' vs CSV '{exp_dept}'")
        if actual_position != exp_position:
            errors.append(f"positionName: DB '{actual_position}' vs CSV '{exp_position}'")
        if actual_dept_code != exp_dept_code:
            errors.append(f"deptCode: DB '{actual_dept_code}' vs CSV '{exp_dept_code}'")
        
        if errors:
            activities_errors.append({
                'doc_id': row['id'],
                'name': name,
                'source': source,
                'errors': errors
            })
        else:
            activities_ok += 1

print("=" * 60)
print("Activities ê²€ì¦ ê²°ê³¼")
print("=" * 60)
print(f"ê²€ì‚¬í•œ activity ìˆ˜: {activities_total}ê±´")
print(f"âœ… ì •ìƒ: {activities_ok}ê±´")
print(f"âŒ ì˜¤ë¥˜: {len(activities_errors)}ê±´")

# Unknown ì¼€ì´ìŠ¤ ì¶œë ¥
unique_unknown_act = set([u['name'] for u in unknown_activities])
print(f"\nâš ï¸  CSVì— ì—†ëŠ” ê²°ì¬ì (unknown): {len(unique_unknown_act)}ëª…")
if unique_unknown_act:
    print(f"   ì´ë¦„ ëª©ë¡: {unique_unknown_act}")

if activities_errors:
    print("\n[ì˜¤ë¥˜ ìƒ˜í”Œ (ìµœëŒ€ 10ê°œ)]")
    for err in activities_errors[:10]:
        print(f"  ë¬¸ì„œ ID {err['doc_id']}: {err['name']} ({err['source']})")
        for e in err['errors']:
            print(f"    - {e}")
```

```python
# ===========================================
# 5. Referrers ê²€ì¦ + Unknown ì§‘ê³„
# ===========================================
referrers_errors = []
referrers_ok = 0
referrers_total = 0
unknown_referrers = []  # Unknown ì¼€ì´ìŠ¤ ì§‘ê³„

for _, row in df_db.iterrows():
    if not row['referrers'] or pd.isna(row['referrers']) or row['referrers'] == '[]':
        continue
    
    try:
        referrers = json.loads(row['referrers'])
    except:
        continue
    
    for ref in referrers:
        referrers_total += 1
        name = str(ref.get('name', '')).strip()
        expected, source = get_expected_info(name)
        
        # Unknown ì¼€ì´ìŠ¤ ì§‘ê³„
        if source == 'unknown':
            unknown_referrers.append({'doc_id': row['id'], 'name': name})
        
        # ì‹¤ì œ ê°’ (referrersëŠ” name, emailIdë§Œ ìˆìŒ)
        actual_email = safe_str(ref.get('emailId', ''))
        
        # ì˜ˆìƒ ê°’
        exp_email = safe_str(expected['emailId'])
        
        # emailIdë§Œ ë¹„êµ
        if actual_email != exp_email:
            referrers_errors.append({
                'doc_id': row['id'],
                'name': name,
                'source': source,
                'error': f"emailId: DB '{actual_email}' vs CSV '{exp_email}'"
            })
        else:
            referrers_ok += 1

print("=" * 60)
print("Referrers ê²€ì¦ ê²°ê³¼")
print("=" * 60)
print(f"ê²€ì‚¬í•œ referrer ìˆ˜: {referrers_total}ê±´")
print(f"âœ… ì •ìƒ: {referrers_ok}ê±´")
print(f"âŒ ì˜¤ë¥˜: {len(referrers_errors)}ê±´")

# Unknown ì¼€ì´ìŠ¤ ì¶œë ¥
unique_unknown_ref = set([u['name'] for u in unknown_referrers])
print(f"\nâš ï¸  CSVì— ì—†ëŠ” ì°¸ì¡°ì (unknown): {len(unique_unknown_ref)}ëª…")
if unique_unknown_ref:
    print(f"   ì´ë¦„ ëª©ë¡: {unique_unknown_ref}")

if referrers_errors:
    print("\n[ì˜¤ë¥˜ ìƒ˜í”Œ (ìµœëŒ€ 10ê°œ)]")
    for err in referrers_errors[:10]:
        print(f"  ë¬¸ì„œ ID {err['doc_id']}: {err['name']} ({err['source']})")
        print(f"    - {err['error']}")
```

```python
# ===========================================
# 6. ëœë¤ ìƒ˜í”Œ ìƒì„¸ í™•ì¸
# ===========================================
import random

print("=" * 60)
print("ëœë¤ ìƒ˜í”Œ ìƒì„¸ í™•ì¸ (5ê±´)")
print("=" * 60)

sample_ids = random.sample(list(df_db['id']), min(5, len(df_db)))

for doc_id in sample_ids:
    row = df_db[df_db['id'] == doc_id].iloc[0]
    print(f"\n[ë¬¸ì„œ ID: {doc_id}]")
    print(f"  drafter_name: {row['drafter_name']}")
    print(f"  drafter_email: {row['drafter_email']}")
    print(f"  drafter_dept: {row['drafter_dept']}")
    print(f"  drafter_position: {row['drafter_position']}")
    print(f"  drafter_dept_code: {row['drafter_dept_code']}")
    
    # CSV ì˜ˆìƒê°’
    expected, source = get_expected_info(row['drafter_name'])
    print(f"  [CSV ì˜ˆìƒê°’ - {source}]")
    print(f"    email: {expected['emailId']}, dept: {expected['deptName']}, position: {expected['positionName']}, dept_code: {expected['deptCode']}")
    
    # activities ì¼ë¶€
    if row['activities'] and pd.notna(row['activities']):
        try:
            acts = json.loads(row['activities'])
            print(f"  activities ({len(acts)}ëª…):")
            for act in acts[:3]:
                name = act.get('name', '')
                exp, src = get_expected_info(name)
                print(f"    - {name}: email={act.get('emailId')}, dept={act.get('deptName')}, position={act.get('positionName')}, deptCode={act.get('deptCode')}")
                print(f"      [CSV {src}]: email={exp['emailId']}, dept={exp['deptName']}, position={exp['positionName']}, deptCode={exp['deptCode']}")
        except:
            print("  activities: íŒŒì‹± ì˜¤ë¥˜")
    
    # referrers ì¼ë¶€
    if row['referrers'] and pd.notna(row['referrers']) and row['referrers'] != '[]':
        try:
            refs = json.loads(row['referrers'])
            print(f"  referrers ({len(refs)}ëª…):")
            for ref in refs[:3]:
                name = ref.get('name', '')
                exp, src = get_expected_info(name)
                print(f"    - {name}: email={ref.get('emailId')}")
                print(f"      [CSV {src}]: email={exp['emailId']}")
        except:
            print("  referrers: íŒŒì‹± ì˜¤ë¥˜")
```

```python
# ===========================================
# 7. ìµœì¢… ìš”ì•½
# ===========================================
print("=" * 60)
print("ìµœì¢… ê²€ì¦ ìš”ì•½")
print("=" * 60)

print(f"\n[Drafter ê²€ì¦]")
print(f"  ì´ ë¬¸ì„œ: {len(df_db)}ê±´")
print(f"  âœ… ì •ìƒ: {drafter_ok}ê±´")
print(f"  âŒ ì˜¤ë¥˜: {len(drafter_errors)}ê±´")
print(f"  âš ï¸  Unknown: {len(unique_unknown)}ëª…")

print(f"\n[Activities ê²€ì¦]")
print(f"  ì´ activity: {activities_total}ê±´")
print(f"  âœ… ì •ìƒ: {activities_ok}ê±´")
print(f"  âŒ ì˜¤ë¥˜: {len(activities_errors)}ê±´")
print(f"  âš ï¸  Unknown: {len(unique_unknown_act)}ëª…")

print(f"\n[Referrers ê²€ì¦]")
print(f"  ì´ referrer: {referrers_total}ê±´")
print(f"  âœ… ì •ìƒ: {referrers_ok}ê±´")
print(f"  âŒ ì˜¤ë¥˜: {len(referrers_errors)}ê±´")
print(f"  âš ï¸  Unknown: {len(unique_unknown_ref)}ëª…")

# ì „ì²´ Unknown í†µí•©
all_unknown = unique_unknown | unique_unknown_act | unique_unknown_ref
print(f"\n[Unknown í†µí•©]")
print(f"  CSVì— ì—†ëŠ” ì´ë¦„ (ì „ì²´): {len(all_unknown)}ëª…")
if all_unknown:
    print(f"  ì´ë¦„ ëª©ë¡: {all_unknown}")

print("\n" + "=" * 60)
total_errors = len(drafter_errors) + len(activities_errors) + len(referrers_errors)
if total_errors == 0 and len(all_unknown) == 0:
    print("ğŸ‰ ëª¨ë“  ê²€ì¦ í†µê³¼! DBì— ì •ìƒì ìœ¼ë¡œ ë°˜ì˜ë˜ì—ˆìŠµë‹ˆë‹¤.")
elif total_errors == 0 and len(all_unknown) > 0:
    print("âœ… ì˜¤ë¥˜ëŠ” ì—†ì§€ë§Œ, CSVì— ì—†ëŠ” ì´ë¦„ì´ ìˆìŠµë‹ˆë‹¤. í™•ì¸ì´ í•„ìš”í•©ë‹ˆë‹¤.")
else:
    print(f"âš ï¸ ì˜¤ë¥˜ê°€ {total_errors}ê±´ ìˆìŠµë‹ˆë‹¤. ìœ„ ê²°ê³¼ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
print("=" * 60)
```

```python
# ===========================================
# 8. (ì„ íƒ) ì˜¤ë¥˜ ìƒì„¸ ë‚´ë³´ë‚´ê¸°
# ===========================================
# ì˜¤ë¥˜ê°€ ìˆì„ ê²½ìš° CSVë¡œ ë‚´ë³´ë‚´ê¸°

export_errors = False  # Trueë¡œ ë³€ê²½í•˜ë©´ ì˜¤ë¥˜ ëª©ë¡ì„ CSVë¡œ ì €ì¥

if export_errors:
    import os
    output_dir = r"C:\Users\LEEJUHWAN\Desktop\ì• ë‹ˆíŒŒì´ë¸Œ\ë°±ì—…"
    
    if drafter_errors:
        df_drafter_err = pd.DataFrame(drafter_errors)
        df_drafter_err['errors'] = df_drafter_err['errors'].apply(lambda x: ' | '.join(x))
        df_drafter_err.to_csv(os.path.join(output_dir, 'drafter_errors.csv'), index=False, encoding='utf-8-sig')
        print(f"drafter_errors.csv ì €ì¥ ì™„ë£Œ ({len(drafter_errors)}ê±´)")
    
    if activities_errors:
        df_act_err = pd.DataFrame(activities_errors)
        df_act_err['errors'] = df_act_err['errors'].apply(lambda x: ' | '.join(x))
        df_act_err.to_csv(os.path.join(output_dir, 'activities_errors.csv'), index=False, encoding='utf-8-sig')
        print(f"activities_errors.csv ì €ì¥ ì™„ë£Œ ({len(activities_errors)}ê±´)")
    
    if referrers_errors:
        df_ref_err = pd.DataFrame(referrers_errors)
        df_ref_err.to_csv(os.path.join(output_dir, 'referrers_errors.csv'), index=False, encoding='utf-8-sig')
        print(f"referrers_errors.csv ì €ì¥ ì™„ë£Œ ({len(referrers_errors)}ê±´)")
    
    if all_unknown:
        df_unknown = pd.DataFrame({'unknown_names': list(all_unknown)})
        df_unknown.to_csv(os.path.join(output_dir, 'unknown_names.csv'), index=False, encoding='utf-8-sig')
        print(f"unknown_names.csv ì €ì¥ ì™„ë£Œ ({len(all_unknown)}ê±´)")
    
    if not (drafter_errors or activities_errors or referrers_errors or all_unknown):
        print("ë‚´ë³´ë‚¼ ì˜¤ë¥˜ê°€ ì—†ìŠµë‹ˆë‹¤.")
else:
    print("ì˜¤ë¥˜ ë‚´ë³´ë‚´ê¸° ë¹„í™œì„±í™” ìƒíƒœì…ë‹ˆë‹¤.")
    print("export_errors = True ë¡œ ë³€ê²½í•˜ë©´ ì˜¤ë¥˜ ëª©ë¡ì„ CSVë¡œ ì €ì¥í•©ë‹ˆë‹¤.")
```
