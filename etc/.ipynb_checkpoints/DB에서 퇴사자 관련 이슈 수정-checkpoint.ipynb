```python
# ===========================================
# 1. DB에서 의심 케이스 조회
# ===========================================
import pymysql
import pandas as pd

conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    password='1234',
    database='any_approval',
    charset='utf8mb4'
)

query = """
SELECT id, drafter_name, drafter_email, drafter_dept, drafter_dept_code, drafter_position
FROM documents
WHERE drafter_email = 'master'
  AND (drafter_dept_code <> '' OR drafter_position <> '' OR drafter_dept <> '')
"""

df_suspect = pd.read_sql(query, conn)
conn.close()

print(f"의심 케이스: {len(df_suspect)}건")
print(df_suspect.head(10))
```

```python
# ===========================================
# 2. 퇴사자 명단 로드
# ===========================================
resigned_path = r"C:\Users\LEEJUHWAN\Desktop\애니파이브\백업\퇴사자명단.csv"
df_resigned = pd.read_csv(resigned_path, encoding='utf-8-sig')
resigned_set = set(df_resigned['퇴사자명'].str.strip())

print(f"퇴사자 수: {len(resigned_set)}")
```

```python
# ===========================================
# 3. 퇴사자 여부 확인
# ===========================================
df_suspect['is_resigned'] = df_suspect['drafter_name'].apply(lambda x: x.strip() in resigned_set)

# 퇴사자 O
df_ok = df_suspect[df_suspect['is_resigned'] == True]
# 퇴사자 X (재직자인데 master로 되어있음 → 문제!)
df_problem = df_suspect[df_suspect['is_resigned'] == False]

print(f"퇴사자 맞음 (정상): {len(df_ok)}건")
print(f"퇴사자 아님 (문제!): {len(df_problem)}건")
```

```python
# ===========================================
# 4. 문제 케이스 확인
# ===========================================
if len(df_problem) > 0:
    print("=== 재직자인데 master로 되어있는 케이스 ===")
    print(df_problem[['id', 'drafter_name', 'drafter_email', 'drafter_dept', 'drafter_position']])
else:
    print("문제 케이스 없음! 202건 모두 퇴사자입니다.")
```
