```python
# ===========================================
# 1. ì¬ì§ì & í‡´ì‚¬ì CSV ë¡œë“œ
# ===========================================
import pymysql
import pandas as pd
import json

# âš ï¸ ê²½ë¡œ ìˆ˜ì • í•„ìš”!
in_path = r"C:\Users\LEEJUHWAN\Desktop\ì• ë‹ˆíŒŒì´ë¸Œ\ë°±ì—…\in_employee.csv"
out_path = r"C:\Users\LEEJUHWAN\Desktop\ì• ë‹ˆíŒŒì´ë¸Œ\ë°±ì—…\out_employee.csv"

df_in = pd.read_csv(in_path, encoding='utf-8-sig')
df_out = pd.read_csv(out_path, encoding='utf-8-sig')

# ì¬ì§ì ë”•ì…”ë„ˆë¦¬: {ì‚¬ì›ëª…: {empNo, deptCode}}
in_dict = {}
for _, row in df_in.iterrows():
    name = str(row['ì‚¬ì›ëª…']).strip()
    emp_no = str(row['ì‚¬ì›ë²ˆí˜¸']).strip() if pd.notna(row['ì‚¬ì›ë²ˆí˜¸']) else ''
    dept_code = str(row['ë¶€ì„œì½”ë“œ']).strip() if pd.notna(row['ë¶€ì„œì½”ë“œ']) else ''
    # nan ë¬¸ìì—´ ì²˜ë¦¬
    emp_no = '' if emp_no == 'nan' else emp_no
    dept_code = '' if dept_code == 'nan' else dept_code
    in_dict[name] = {
        'empNo': emp_no,
        'deptCode': dept_code
    }

# í‡´ì‚¬ì ë”•ì…”ë„ˆë¦¬: {ì‚¬ì›ëª…: {empNo, deptCode}}
out_dict = {}
for _, row in df_out.iterrows():
    name = str(row['ì‚¬ì›ëª…']).strip()
    emp_no = str(row['ì‚¬ì›ë²ˆí˜¸']).strip() if pd.notna(row['ì‚¬ì›ë²ˆí˜¸']) else ''
    dept_code = str(row['ë¶€ì„œì½”ë“œ']).strip() if pd.notna(row['ë¶€ì„œì½”ë“œ']) else ''
    # nan ë¬¸ìì—´ ì²˜ë¦¬
    emp_no = '' if emp_no == 'nan' else emp_no
    dept_code = '' if dept_code == 'nan' else dept_code
    out_dict[name] = {
        'empNo': emp_no,
        'deptCode': dept_code
    }

print(f"ì¬ì§ì: {len(in_dict)}ëª…")
print(f"í‡´ì‚¬ì: {len(out_dict)}ëª…")
print(f"ì¬ì§ì ìƒ˜í”Œ: {list(in_dict.items())[:2]}")
print(f"í‡´ì‚¬ì ìƒ˜í”Œ: {list(out_dict.items())[:2]}")
```

```python
# ===========================================
# 2. documents í…Œì´ë¸” ì¡°íšŒ (referrers)
# ===========================================
conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    password='1234',
    database='any_approval',
    charset='utf8mb4'
)

query = "SELECT id, referrers FROM documents"
df_docs = pd.read_sql(query, conn)
conn.close()

print(f"documents ì´ ê±´ìˆ˜: {len(df_docs)}")
print(df_docs.head())
```

```python
# ===========================================
# 3. ì—…ë°ì´íŠ¸ ë°ì´í„° ìƒì„±
# ===========================================
def get_person_info(name):
    """ì‚¬ëŒ ì •ë³´ ë°˜í™˜ (ì¬ì§ì ìš°ì„ , ê·¸ë‹¤ìŒ í‡´ì‚¬ì)"""
    name = str(name).strip() if name else ''
    
    # 1ìˆœìœ„: ì¬ì§ì
    if name in in_dict:
        info = in_dict[name]
        return {
            'empNo': info['empNo'],
            'deptCode': info['deptCode'],
            'source': 'in'
        }
    
    # 2ìˆœìœ„: í‡´ì‚¬ì
    if name in out_dict:
        info = out_dict[name]
        return {
            'empNo': info['empNo'],
            'deptCode': info['deptCode'],
            'source': 'out'
        }
    
    # ë‘˜ ë‹¤ ì—†ìœ¼ë©´ (ë¯¸ë“±ë¡ì)
    return {
        'empNo': '',
        'deptCode': '',
        'source': 'unknown'
    }


def process_referrers(referrers_str):
    """referrers JSON ì²˜ë¦¬"""
    if not referrers_str or pd.isna(referrers_str):
        return None, []
    
    try:
        referrers = json.loads(referrers_str)
    except json.JSONDecodeError:
        return None, []
    
    unknown_names = []
    
    for ref in referrers:
        name = ref.get('name', '').strip()
        info = get_person_info(name)
        
        ref['empNo'] = info['empNo']
        ref['deptCode'] = info['deptCode']
        
        if info['source'] == 'unknown' and name:
            unknown_names.append(name)
    
    return json.dumps(referrers, ensure_ascii=False), unknown_names


# ì—…ë°ì´íŠ¸ ë°ì´í„° ìƒì„±
update_data = []
unknown_referrers = []

for idx, row in df_docs.iterrows():
    new_referrers, unknowns = process_referrers(row['referrers'])
    unknown_referrers.extend(unknowns)
    
    update_data.append({
        'id': row['id'],
        'referrers': new_referrers
    })

df_update = pd.DataFrame(update_data)
print(f"ì—…ë°ì´íŠ¸ ëŒ€ìƒ: {len(df_update)}ê±´")
print(df_update.head(10))

# ë¯¸ë“±ë¡ì í™•ì¸
unknown_unique = list(set(unknown_referrers))
print(f"\nâš ï¸ ì¬ì§ì/í‡´ì‚¬ì ëª©ë¡ ë‘˜ ë‹¤ ì—†ëŠ” ì‚¬ëŒ: {len(unknown_unique)}ëª…")
if unknown_unique:
    print(unknown_unique[:30])
```

```python
# ===========================================
# 4. ê²€ì¦ (ì—…ë°ì´íŠ¸ ì „ ë¯¸ë¦¬ë³´ê¸°)
# ===========================================
print("=" * 60)
print("ê²€ì¦")
print("=" * 60)

# referrersê°€ ìˆëŠ” ë¬¸ì„œ ìˆ˜
has_referrers = df_update[df_update['referrers'].notna()].shape[0]
no_referrers = df_update[df_update['referrers'].isna()].shape[0]
print(f"referrers ìˆëŠ” ë¬¸ì„œ: {has_referrers}ê±´")
print(f"referrers ì—†ëŠ” ë¬¸ì„œ: {no_referrers}ê±´")

# referrers ë‚´ ì‚¬ëŒ í†µê³„
total_refs = 0
in_count = 0
out_count = 0
unknown_count = 0

for _, row in df_update.iterrows():
    if not row['referrers'] or pd.isna(row['referrers']):
        continue
    try:
        refs = json.loads(row['referrers'])
        for ref in refs:
            total_refs += 1
            name = ref.get('name', '').strip()
            if name in in_dict:
                in_count += 1
            elif name in out_dict:
                out_count += 1
            else:
                unknown_count += 1
    except:
        continue

print(f"\n[referrers ë‚´ ì‚¬ëŒ í†µê³„]")
print(f"  ì´ ì°¸ì¡°ì ìˆ˜: {total_refs}ëª…")
print(f"  ì¬ì§ì: {in_count}ëª…")
print(f"  í‡´ì‚¬ì: {out_count}ëª…")
print(f"  ë¯¸ë“±ë¡ì: {unknown_count}ëª…")
```

```python
# ===========================================
# 5. ìƒ˜í”Œ í™•ì¸
# ===========================================
print("=" * 60)
print("ìƒ˜í”Œ í™•ì¸ (referrers ìˆëŠ” ë¬¸ì„œ 5ê±´)")
print("=" * 60)

sample_count = 0
for _, row in df_update.iterrows():
    if not row['referrers'] or pd.isna(row['referrers']):
        continue
    
    try:
        refs = json.loads(row['referrers'])
        if len(refs) > 0:
            print(f"\n[ë¬¸ì„œ ID: {row['id']}]")
            for ref in refs:
                name = ref.get('name', '')
                info = get_person_info(name)
                print(f"  {name}: empNo={ref.get('empNo')}, deptCode={ref.get('deptCode')} ({info['source']})")
            sample_count += 1
            if sample_count >= 5:
                break
    except:
        continue
```

```python
# ===========================================
# 6. ì‹¤ì œ DB ì—…ë°ì´íŠ¸ (í™•ì¸ í›„ ì‹¤í–‰!)
# ===========================================
# âš ï¸ ìœ„ ê²€ì¦ ê²°ê³¼ í™•ì¸ í›„ ì‹¤í–‰!

conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    password='1234',
    database='any_approval',
    charset='utf8mb4'
)
cursor = conn.cursor()

update_count = 0
error_count = 0

for idx, row in df_update.iterrows():
    try:
        query = """UPDATE documents 
        SET referrers = %s
        WHERE id = %s"""
        
        cursor.execute(query, (
            row['referrers'],
            row['id']
        ))
        update_count += 1
        
        if update_count % 1000 == 0:
            print(f"ì§„í–‰ ì¤‘... {update_count}/{len(df_update)}")
            
    except Exception as e:
        error_count += 1
        print(f"ì—ëŸ¬ - ID {row['id']}: {e}")

conn.commit()
cursor.close()
conn.close()

print(f"\nì™„ë£Œ! ì—…ë°ì´íŠ¸: {update_count}ê±´, ì—ëŸ¬: {error_count}ê±´")
```

```python
# ===========================================
# 7. DB ì—…ë°ì´íŠ¸ í›„ ê²€ì¦
# ===========================================
conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    password='1234',
    database='any_approval',
    charset='utf8mb4'
)

query = "SELECT id, referrers FROM documents"
df_db = pd.read_sql(query, conn)
conn.close()

# ê²€ì¦
referrers_errors = []
referrers_ok = 0
referrers_total = 0

for _, row in df_db.iterrows():
    if not row['referrers'] or pd.isna(row['referrers']):
        continue
    
    try:
        refs = json.loads(row['referrers'])
    except:
        continue
    
    for ref in refs:
        referrers_total += 1
        name = ref.get('name', '').strip()
        info = get_person_info(name)
        
        # ì‹¤ì œ ê°’
        actual_empNo = str(ref.get('empNo', '')).strip()
        actual_deptCode = str(ref.get('deptCode', '')).strip()
        
        # ì˜ˆìƒ ê°’
        exp_empNo = info['empNo']
        exp_deptCode = info['deptCode']
        
        # ë¹„êµ
        errors = []
        if actual_empNo != exp_empNo:
            errors.append(f"empNo: DB '{actual_empNo}' vs CSV '{exp_empNo}'")
        if actual_deptCode != exp_deptCode:
            errors.append(f"deptCode: DB '{actual_deptCode}' vs CSV '{exp_deptCode}'")
        
        if errors:
            referrers_errors.append({
                'doc_id': row['id'],
                'name': name,
                'source': info['source'],
                'errors': errors
            })
        else:
            referrers_ok += 1

print("=" * 60)
print("Referres ê²€ì¦ ê²°ê³¼")
print("=" * 60)
print(f"ê²€ì‚¬í•œ ì°¸ì¡°ì ìˆ˜: {referrers_total}ê±´")
print(f"âœ… ì •ìƒ: {referrers_ok}ê±´")
print(f"âŒ ì˜¤ë¥˜: {len(referrers_errors)}ê±´")

if referrers_errors:
    print("\n[ì˜¤ë¥˜ ìƒ˜í”Œ (ìµœëŒ€ 10ê°œ)]")
    for err in referrers_errors[:10]:
        print(f"  ë¬¸ì„œ ID {err['doc_id']}: {err['name']} ({err['source']})")
        for e in err['errors']:
            print(f"    - {e}")
else:
    print("\nğŸ‰ ëª¨ë“  ê²€ì¦ í†µê³¼! referrersê°€ ì •ìƒì ìœ¼ë¡œ ë°˜ì˜ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
