# Empty Path 데이터 조회 및 CSV 저장
path=""인 85건의 year, title, doc_num 조회

```python
import pymysql
import pandas as pd
from datetime import datetime
```

```python
# DB 접속 정보
DB_CONFIG = {
    'host': 'localhost',
    'user': 'root',
    'password': '1234',
    'database': 'any_approval',
    'charset': 'utf8mb4'
}
```

```python
# DB 연결
conn = pymysql.connect(**DB_CONFIG)
print("DB 연결 성공")
```

```python
# Empty path 데이터 조회
query = """
SELECT source_id, year, title, doc_num, attaches 
FROM documents 
WHERE attaches LIKE '%"path": ""%' OR attaches LIKE '%"path":""%'
ORDER BY year, source_id
"""

df = pd.read_sql(query, conn)
print(f"조회 결과: {len(df)}건")
```

```python
# 데이터 미리보기
df.head(10)
```

```python
# 연도별 분포
print("연도별 분포:")
print(df['year'].value_counts().sort_index())
```

```python
# CSV 저장 (attaches 컬럼 포함)
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
filename = f'empty_path_documents_{timestamp}.csv'

df.to_csv(filename, index=False, encoding='utf-8-sig')
print(f"저장 완료: {filename}")
```

```python
# 간단한 버전 (source_id, year, title, doc_num만)
df_simple = df[['source_id', 'year', 'title', 'doc_num']]
filename_simple = f'empty_path_documents_simple_{timestamp}.csv'

df_simple.to_csv(filename_simple, index=False, encoding='utf-8-sig')
print(f"저장 완료: {filename_simple}")
```

```python
# 연결 종료
conn.close()
print("DB 연결 종료")
```
