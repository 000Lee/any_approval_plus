## 1. ì„¤ì •

```python
import json
import pymysql
import pandas as pd

# DB ì„¤ì •
DB_CONFIG = {
    'host': 'localhost',
    'user': 'root',
    'password': '1234',
    'database': 'any_approval',
    'charset': 'utf8mb4'
}

# end_year â†’ approval_data í…Œì´ë¸” ë§¤í•‘
APPROVAL_TABLE_MAPPING = {
    2010: 'approval_data_2010',
    2015: 'approval_data_2015',
    2020: 'approval_data_2020',
    2025: 'approval_data_2025'
}

# status â†’ actionLogType ë§¤í•‘
STATUS_MAPPING = {
    'ê¸°ì•ˆ': 'DRAFT',
    'ìŠ¹ì¸': 'APPROVAL',
    'í•©ì˜': 'AGREEMENT',
    'í•©ì˜(ëŒ€ê²°)': 'AGREEMENT',
    'ë°˜ë ¤': 'RETURN'
}

print('ì„¤ì • ì™„ë£Œ!')
```

## 2. documents í…Œì´ë¸” ì¡°íšŒ

```python
def get_documents():
    """documents í…Œì´ë¸”ì—ì„œ source_id, end_year, activities ì¡°íšŒ"""
    conn = pymysql.connect(**DB_CONFIG)
    
    try:
        query = """
        SELECT source_id, end_year, title, activities
        FROM documents
        ORDER BY end_year, source_id
        """
        df = pd.read_sql(query, conn)
        print(f'âœ… documents í…Œì´ë¸”: {len(df)}ê±´ ì¡°íšŒ')
        print(df.groupby('end_year').size())
        return df
    finally:
        conn.close()

documents_df = get_documents()
```

## 3. approval_data í…Œì´ë¸”ë“¤ ì¡°íšŒ

```python
def get_approval_data():
    """ëª¨ë“  approval_data í…Œì´ë¸” ì¡°íšŒ í›„ í•©ì¹˜ê¸°"""
    conn = pymysql.connect(**DB_CONFIG)
    all_data = []
    
    try:
        for end_year, table_name in APPROVAL_TABLE_MAPPING.items():
            query = f"""
            SELECT 
                document_id,
                sequence,
                status,
                approver,
                {end_year} as end_year
            FROM {table_name}
            ORDER BY document_id, sequence
            """
            df = pd.read_sql(query, conn)
            print(f'âœ… {table_name}: {len(df)}ê±´')
            all_data.append(df)
        
        combined_df = pd.concat(all_data, ignore_index=True)
        print(f'\nğŸ“Š ì „ì²´ approval_data: {len(combined_df)}ê±´')
        return combined_df
    
    finally:
        conn.close()

approval_df = get_approval_data()
```

## 4. ë¶ˆì¼ì¹˜ ê²€ì¦ ì‹¤í–‰

```python
def verify_action_types(documents_df, approval_df):
    mismatches = []
    
    # approval_dataë¥¼ dictë¡œ ë³€í™˜ - ê°’ì„ ë¦¬ìŠ¤íŠ¸ë¡œ!
    approval_dict = {}
    for _, row in approval_df.iterrows():
        approver_parts = str(row['approver']).split() if row['approver'] else []
        if len(approver_parts) > 1:
            approver_name = ' '.join(approver_parts[:-1])
        else:
            approver_name = approver_parts[0] if approver_parts else ''
        
        key = (str(row['document_id']), row['end_year'], approver_name)
        
        # â­ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥ (í•œ ì‚¬ëŒì´ ì—¬ëŸ¬ ë²ˆ ê²°ì¬ ê°€ëŠ¥)
        if key not in approval_dict:
            approval_dict[key] = []
        approval_dict[key].append(row['status'])
    
    print(f'approval_dict ìƒì„± ì™„ë£Œ: {len(approval_dict)}ê±´')
    
    total = len(documents_df)
    matched_count = 0
    not_found_count = 0
    
    for idx, row in documents_df.iterrows():
        source_id = str(row['source_id'])
        end_year = row['end_year']
        title = row['title']
        
        if (idx + 1) % 1000 == 0:
            print(f'ì§„í–‰: {idx + 1}/{total}')
        
        try:
            activities = json.loads(row['activities'], strict=False) if row['activities'] else []
        except:
            continue
        
        for act in activities:
            name = act.get('name', '')
            db_type = act.get('actionLogType', '')
            
            if not name:
                continue
            
            key = (source_id, end_year, name)
            
            if key not in approval_dict:
                not_found_count += 1
                continue
            
            matched_count += 1
            status_list = approval_dict[key]  # â­ ë¦¬ìŠ¤íŠ¸ë¡œ ê°€ì ¸ì˜´
            
            # â­ status ë¦¬ìŠ¤íŠ¸ â†’ expected_type ë¦¬ìŠ¤íŠ¸
            expected_types = [STATUS_MAPPING.get(s, '') for s in status_list]
            
            # â­ db_typeì´ ë¦¬ìŠ¤íŠ¸ì— ìˆìœ¼ë©´ OK
            if db_type not in expected_types:
                mismatches.append({
                    'source_id': source_id,
                    'end_year': end_year,
                    'title': title,
                    'name': name,
                    'approval_status_list': status_list,
                    'expected_types': expected_types,
                    'db_type': db_type
                })
    
    print(f'\nâœ… ê²€ì¦ ì™„ë£Œ')
    print(f'   ë§¤ì¹­ ì„±ê³µ: {matched_count}ê±´')
    print(f'   ë§¤ì¹­ ì‹¤íŒ¨ (approval_dataì— ì—†ìŒ): {not_found_count}ê±´')
    
    return mismatches

# ê²€ì¦ ì‹¤í–‰
print('ê²€ì¦ ì‹œì‘...')
mismatches = verify_action_types(documents_df, approval_df)
print(f'\nğŸ”´ ë¶ˆì¼ì¹˜ ë°œê²¬: {len(mismatches)}ê±´')
```

```python
# 1. ë¨¼ì € not_matched_df ìƒˆë¡œ ë§Œë“¤ê¸° (ìˆ˜ì •ëœ ë¡œì§ìœ¼ë¡œ)
def find_not_matched(documents_df, approval_df):
    approval_dict = {}
    for _, row in approval_df.iterrows():
        approver_parts = str(row['approver']).split() if row['approver'] else []
        if len(approver_parts) > 1:
            approver_name = ' '.join(approver_parts[:-1])
        else:
            approver_name = approver_parts[0] if approver_parts else ''
        
        key = (str(row['document_id']), row['end_year'], approver_name)
        approval_dict[key] = row['status']
    
    not_matched = []
    for idx, row in documents_df.iterrows():
        source_id = str(row['source_id'])
        end_year = row['end_year']
        
        try:
            activities = json.loads(row['activities']) if row['activities'] else []
        except:
            continue
        
        for act in activities:
            name = act.get('name', '')
            if not name:
                continue
            
            key = (source_id, end_year, name)
            if key not in approval_dict:
                not_matched.append({
                    'source_id': source_id,
                    'end_year': end_year,
                    'name': name,
                    'title': row['title']
                })
    
    return not_matched

not_matched = find_not_matched(documents_df, approval_df)
not_matched_df = pd.DataFrame(not_matched)

# 2. ê·¸ ë‹¤ìŒ ì›ì¸ ë¶„ì„
doc_not_exists, name_not_found = analyze_not_matched(not_matched_df, approval_df)

print(f'ğŸ“Š ë§¤ì¹­ ì‹¤íŒ¨ ì›ì¸ ë¶„ì„')
print(f'=' * 50)
print(f'1. ë¬¸ì„œ ìì²´ê°€ approval_dataì— ì—†ìŒ: {len(doc_not_exists)}ê±´')
print(f'2. ë¬¸ì„œëŠ” ìˆëŠ”ë° ì´ë¦„ ë§¤ì¹­ ì‹¤íŒ¨: {len(name_not_found)}ê±´')

if name_not_found:
    print(f'\nğŸ“‹ ì´ë¦„ ë§¤ì¹­ ì‹¤íŒ¨ ìƒì„¸:')
    for i, item in enumerate(name_not_found[:10], 1):
        print(f"\n[{i}] source_id: {item['source_id']}, end_year: {item['end_year']}")
        print(f"    documents name: {item['name']}")
        print(f"    approval_data approvers: {item['approval_approvers']}")
```

## 7. (ì„ íƒ) ë¶ˆì¼ì¹˜ DB ì—…ë°ì´íŠ¸

```python
# âš ï¸ ì‹¤í–‰ ì „ ë°˜ë“œì‹œ ë°±ì—…í•˜ì„¸ìš”!
UPDATE_DB = True  # Trueë¡œ ë°”ê¾¸ë©´ ì‹¤í–‰ë¨

if UPDATE_DB and mismatches:
    conn = pymysql.connect(**DB_CONFIG)
    cursor = conn.cursor()
    
    from collections import defaultdict
    doc_updates = defaultdict(list)
    
    for m in mismatches:
        key = (m['source_id'], m['end_year'])
        # expected_types ë¦¬ìŠ¤íŠ¸ì—ì„œ ì²« ë²ˆì§¸ ìœ íš¨í•œ ê°’ ì‚¬ìš©
        expected_type = next((t for t in m['expected_types'] if t), '')
        if expected_type:
            doc_updates[key].append((m['name'], expected_type))
    
    updated = 0
    
    for (source_id, end_year), updates in doc_updates.items():
        cursor.execute(
            "SELECT activities FROM documents WHERE source_id = %s AND end_year = %s",
            (source_id, end_year)
        )
        row = cursor.fetchone()
        if not row:
            continue
        
        activities = json.loads(row[0]) if row[0] else []
        
        modified = False
        for name, expected_type in updates:
            for act in activities:
                if act.get('name') == name:
                    act['actionLogType'] = expected_type
                    act['type'] = expected_type
                    modified = True
                    break
        
        if modified:
            cursor.execute(
                "UPDATE documents SET activities = %s WHERE source_id = %s AND end_year = %s",
                (json.dumps(activities, ensure_ascii=False), source_id, end_year)
            )
            updated += 1
    
    conn.commit()
    cursor.close()
    conn.close()
    
    print(f'âœ… {updated}ê°œ ë¬¸ì„œ ì—…ë°ì´íŠ¸ ì™„ë£Œ (ì´ {len(mismatches)}ê±´ ë¶ˆì¼ì¹˜ ìˆ˜ì •)')
else:
    print('UPDATE_DB = Trueë¡œ ì„¤ì •í•˜ë©´ ë¶ˆì¼ì¹˜ê°€ DBì— ì—…ë°ì´íŠ¸ë©ë‹ˆë‹¤.')
```
